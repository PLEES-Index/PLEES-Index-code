Read the projected population from the UN 5 year age grouped poplulation tables 1950-2100 
From this table we use the country and age specific population (PopTotal)

```python
pop_data = pd.read_csv("../data/input/WPP2022_Population1JanuaryByAge5GroupSex_Medium.csv",
                       usecols=["ISO3_code","Time","AgeGrp","PopTotal"],
                       dtype= {"ISO3_code":str,"Time":str,"AgeGrp":str,"PopTotal":float})
# Select only entries for countries, both sexes, and the years 1975, 2022 and 2080 and the age category 0-4.
pop_data = pop_data[(pop_data.ISO3_code.notnull()) & 
                    (pop_data.Time.isin(["1975", "2010","2080"]))& 
                    (pop_data.AgeGrp != "0-4")]
pop_data.loc[pop_data.AgeGrp == "5-9", "AgeGrp"] = "05-09"
pop_data.loc[pop_data.AgeGrp == "100+", "AgeGrp"] = "99+"

###
# Read the projected population for 2080 from the UN Poplulation tables 2022-2100 
###
pop_data_temp = pd.read_csv("../data/input/WPP2022_Population1JanuaryBySingleAgeSex_Medium_2022-2100.csv",
                       usecols=["ISO3_code","Time","AgeGrp","PopTotal"],
                       dtype= {"ISO3_code":str,"Time":str,"AgeGrp":str,"PopTotal":float})


# Select only entries for countries, both sexes, and the years 2022 and 2080 and the age category 0-4.
pop_data_temp = pop_data_temp[(pop_data_temp.ISO3_code.notnull()) & 
                              (pop_data_temp.Time.isin(["2080"]))& 
                              (pop_data_temp.AgeGrp.isin(["0","1","2","3","4"]))]

# Combine age group 1-4
pop_data_temp.loc[pop_data_temp.AgeGrp != "0", "AgeGrp"] = "01-04"
pop_data_temp = pop_data_temp.groupby(['ISO3_code', 'Time',"AgeGrp"]).sum().reset_index()
pop_data = pd.concat([pop_data,pop_data_temp])
pop_data = pop_data.sort_values(by=["ISO3_code","Time","AgeGrp"])

###
# Read the projected population for 1975 and 2010 from the UN Poplulation tables 1950-2021 
###
pop_data_temp = pd.read_csv("../data/input/WPP2024_Population1JanuaryBySingleAgeSex_Medium_1950-2023.csv",
                       usecols=["ISO3_code","Time","AgeGrp","PopTotal"],
                       dtype= {"ISO3_code":str,"Time":str,"AgeGrp":str,"PopTotal":float})


# Select only entries for countries, both sexes, and the years 2010 and 2080 and the age category 0-4.
pop_data_temp = pop_data_temp[(pop_data_temp.ISO3_code.notnull()) & 
                              (pop_data_temp.Time.isin(["1975","2010"]))& 
                              (pop_data_temp.AgeGrp.isin(["0","1","2","3","4"]))]

# Combine age group 1-4
pop_data_temp.loc[pop_data_temp.AgeGrp != "0", "AgeGrp"] = "01-04"
pop_data_temp = pop_data_temp.groupby(['ISO3_code', 'Time',"AgeGrp"]).sum().reset_index()
pop_data = pd.concat([pop_data,pop_data_temp])
pop_data = pop_data.sort_values(by=["ISO3_code","Time","AgeGrp"])

# Set the country and age group as index
pop_data["PopTotal"] = pop_data["PopTotal"]*1000
pop_data.set_index(["ISO3_code","AgeGrp"], inplace=True)
pop_data
```

# Calculate the life expectancy according to Hauer et al 2021
Here we define the life expectancy calculation:

```python
def le(ehf, data):
    # Add the heat fatalities  to the 
    mor = data["mx"] + ehf["val"]
    
    # Probability of dying
    q = (mor* data["AgeGrpSpan"] / (1+(data["AgeGrpSpan"]-data["ax"])*mor )).to_frame(name="val")
    l = q
    l["val"] = (1 - q["val"])
    l["valminone"]= (l.groupby(["ISO3_code"], as_index=False)["val"].cumprod())
    l["val"] = l.groupby(["ISO3_code"], as_index=False)["valminone"].shift(1)
    l.loc[l.val.isna(), "val"] = 1
    l["valminone"] = l["valminone"]*100000
    l["val"] = l["val"]*100000

    # Person years lived
    pyl = (data["ax"] * l["val"] + (data["AgeGrpSpan"] - data["ax"])*l["valminone"]).to_frame(name="val")

    # Total person years lived
    T = (pyl.iloc[::-1].groupby(["ISO3_code"], as_index=False)["val"].cumsum()).iloc[::-1].to_frame(name="val")
    print(T)
    # Life expectancy
    e = (T["val"] / l["val"]).to_frame(name="val")
    return e
```

Next we calculate the life expectancy for each scenario

```python
# Calculate the life expectancy for 2022 without excess heat deaths
exp_base = le(ehf_blank,mor_data[mor_data["Time"] == "2010"]).reset_index()
exp_table = exp_base[exp_base["AgeGrp"] == "0"]
exp_table = exp_table.rename(columns={"val":"LifeExpNo_00_2010"})
exp_table = exp_table.drop(columns=["AgeGrp"])

# Calculate the life expectancy for 2010
exp_temp = le(ehf_00_2010,mor_data[mor_data["Time"] == "2010"]).reset_index()
exp_table["LifeExp_00_2010"] = exp_temp[exp_temp["AgeGrp"] == "0"]["val"]
exp_table["LE_00_2010"] = 1- (exp_table["LifeExpNo_00_2010"]-exp_table["LifeExp_00_2010"])/10


# Calculate the life expectancy for 1975 without excess heat deaths
exp_temp = le(ehf_blank,mor_data[mor_data["Time"] == "1975"]).reset_index()
exp_table["LifeExpNo_00_1975"] = exp_temp[exp_temp["AgeGrp"] == "0"]["val"]

# Calculate the life expectancy for 1975
exp_temp = le(ehf_00_1975,mor_data[mor_data["Time"] == "1975"]).reset_index()
exp_table["LifeExp_00_1975"] = exp_temp[exp_temp["AgeGrp"] == "0"]["val"]
exp_table["LE_00_1975"] = 1- (exp_table["LifeExpNo_00_1975"]-exp_table["LifeExp_00_1975"])/10


# Calculate the life expectancy for 4.5 2080 without excess heat deaths
exp_temp = le(ehf_blank,mor_data[mor_data["Time"] == "2080"]).reset_index()
exp_table["LifeExpNo_45_2080"] = exp_temp[exp_temp["AgeGrp"] == "0"]["val"]

# Calculate the life expectancy for 4.5 2080
exp_temp = le(ehf_45_2080,mor_data[mor_data["Time"] == "2080"]).reset_index()
exp_table["LifeExp_45_2080"] = exp_temp[exp_temp["AgeGrp"] == "0"]["val"]
exp_table["LE_45_2080"] = 1- (exp_table["LifeExpNo_45_2080"]-exp_table["LifeExp_45_2080"])/10


# Calculate the life expectancy for 8.5 2080 without excess heat deaths
exp_temp = le(ehf_blank,mor_data[mor_data["Time"] == "2080"]).reset_index()
exp_table["LifeExpNo_85_2080"] = exp_temp[exp_temp["AgeGrp"] == "0"]["val"]

# Calculate the life expectancy for 8.5 2080
exp_temp = le(ehf_85_2080,mor_data[mor_data["Time"] == "2080"]).reset_index()
exp_table["LifeExp_85_2080"] = exp_temp[exp_temp["AgeGrp"] == "0"]["val"]
exp_table["LE_85_2080"] = 1- (exp_table["LifeExpNo_85_2080"]-exp_table["LifeExp_85_2080"])/10
```

Next we add all of the necessary columns needed for the rest of the PLEES index.

```python
# Add column for the country names
exp_table["country_name"] = exp_table["ISO3_code"]
exp_table["continent"]    = exp_table["ISO3_code"]

# Read the unified country names and codes file
codes_data = pd.read_csv("../../00_SharedInput/countrycodes.csv",
                         usecols=["country_name","country_code","continent"])

for loc in exp_table["ISO3_code"].unique():
    if loc in codes_data["country_code"].values:
        exp_table.loc[exp_table["ISO3_code"] == loc, "continent"] = codes_data.loc[codes_data["country_code"] == loc, "continent"].values[0]
        exp_table.loc[exp_table["ISO3_code"] == loc, "country_name"] = codes_data.loc[codes_data["country_code"] == loc, "country_name"].values[0]
    else:
        print("Could not match country: ",loc)

exp_table = exp_table.rename(columns={"ISO3_code":"country_code"})
exp_table = exp_table.dropna()
exp_table = exp_table[["country_name","country_code","continent",
          "LifeExpNo_00_1975","LifeExp_00_1975","LE_00_1975",
          "LifeExpNo_00_2010","LifeExp_00_2010","LE_00_2010",
          "LifeExpNo_45_2080","LifeExp_45_2080","LE_45_2080",
          "LifeExpNo_85_2080","LifeExp_85_2080","LE_85_2080"]]
exp_table
```
